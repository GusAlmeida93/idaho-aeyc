```python
import os
import pandas as pd
import requests
```

```python
def parse_data(url, csv=False, columns_headers = True, columns=[] ):
    """
    Parse d
    ata from the url and return a dataframe
    """
    
    headers = {"User-Agent":"Mozilla/5.0"}
    
    response = requests.get(url, headers=headers)

    data = response.text.split('\n')
    
    if columns_headers:
        if csv:
            columns = data[0]
            columns = columns.split(',')
            columns = [i.replace('\r','').replace('"','') for i in columns]
        else:
            columns = data[0]
            columns = columns.split('\t')
            columns = [i.replace('\r','') for i in columns]
    
    if columns_headers:
        
        data = data[1:]
    
    if csv:
        
        data = [i.replace('\r','').replace('"','').split(',') for i in data]
    else:
        
        data = [i.replace('\r','').split('\t') for i in data]
    
    
    
    df = pd.DataFrame(data)
    df = df.replace('', None)
    df = df.dropna(how='all', axis=1)
    df.columns = columns
    df = df.dropna(how='all', axis=0)
    
    return df
```
